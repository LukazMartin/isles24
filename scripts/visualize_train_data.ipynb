```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# change only this
case_id = 3

base = "/home/renku/work/isles24/data-local/train/derivatives"
case = f"sub-stroke{case_id:04d}"
ses = "ses-01"

# file paths
cta_path = f"{base}/{case}/{ses}/{case}_{ses}_space-ncct_cta.nii.gz"
ctp_path = f"{base}/{case}/{ses}/{case}_{ses}_space-ncct_brain-msk.nii.gz"
cow_path = f"{base}/{case}/{ses}/{case}_{ses}_space-ncct_cow-msk.nii.gz"
lvo_path = f"{base}/{case}/{ses}/{case}_{ses}_space-ncct_lvo-msk.nii.gz"

print("Loading case:", case)

cta = nib.load(cta_path).get_fdata()
ctp = nib.load(ctp_path).get_fdata()
cow = nib.load(cow_path).get_fdata()
lvo = nib.load(lvo_path).get_fdata()

print("CTA shape:", cta.shape)

# middle slice
slice_idx = cta.shape[2] // 2

fig, ax = plt.subplots(1,4, figsize=(18,5))

ax[0].imshow(cta[:,:,slice_idx], cmap="gray")
ax[0].set_title("CTA")

ax[1].imshow(ctp[:,:,slice_idx], cmap="gray")
ax[1].set_title("Brain mask")

ax[2].imshow(cow[:,:,slice_idx], cmap="hot")
ax[2].set_title("COW mask")

ax[3].imshow(lvo[:,:,slice_idx], cmap="hot")
ax[3].set_title("LVO mask")

for a in ax:
    a.axis("off")

plt.suptitle(f"{case} | slice {slice_idx}")
plt.show()
```
